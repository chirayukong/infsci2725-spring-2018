```python
import matplotlib.pyplot as plt  # https://matplotlib.org/api/pyplot_summary.html
import pandas as pd  # https://pandas.pydata.org/pandas-docs/stable/
import seaborn as sns  # https://seaborn.pydata.org/api.html
%matplotlib inline
```

```python
print('Read the assignment-01/Batting.csv from github into Pandas DataFrame called df.')
df = pd.read_csv('https://raw.githubusercontent.com/chirayukong/infsci2725-spring-2018/master/assignment-01/Batting.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
print('Sum df by year for team OAK where into oak_year_df.')
oak_year_df = df[(df.teamID == 'OAK')].pivot_table(index='yearID', aggfunc='sum')
_ = oak_year_df.plot()
```

```python
print("Drop Games from oak_year_df because they don't make any sense.")
oak_year_df = oak_year_df.drop('Games', axis=1)
_ = oak_year_df.plot()
```

```python
print('Plot oak_year_df Homeruns vs Walks as a scatterplot with marginal histograms.')
_ = sns.jointplot(data=oak_year_df, x='Walks', y='Homeruns')
```

```python
print('Replace the scatterplot with a joint histogram using hexagonal bin.')
_ = sns.jointplot(data=oak_year_df, x='Walks', y='Homeruns', kind="hex")
```

```python
print('Replace the scatterplots and histograms with density estimates.')
_ = sns.jointplot(data=oak_year_df, x='Walks', y='Homeruns', kind="kde")
```

```python
print('Overlay carpet/rug plot.')
_, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(oak_year_df.Walks,oak_year_df.Homeruns, ax=ax, shade=True)
sns.rugplot(oak_year_df.Walks, color="g", ax=ax)
_ = sns.rugplot(oak_year_df.Homeruns, vertical=True, ax=ax)
```

```python
print('Draw scatterplots with fitting linear regression for joint relationships and histograms for univariate distributions of oak_year_df')
print('This is kind of graphic version of Correlation matrix.')
_ = sns.pairplot(oak_year_df, kind='reg')
```

```python
print('Plot distribution of oak_year_df.Walks (histogram and KDE are default).')
_ = sns.distplot(oak_year_df.Walks)
```

```python
print('Replace KDE with carpet/rug plot.')
_ = sns.distplot(oak_year_df.Walks, kde=False, rug=True)
```

```python
print('Replace histogram with KDE.')
_ = sns.distplot(oak_year_df.Walks, hist=False, rug=True)
```

```python
print('Sum df Homeruns by yearID against teamID where yearID > 2000 into team_homeruns_per_year_df and fill N/A with 0.0.')
team_homeruns_per_year_df = df[df.yearID > 2000].pivot_table(values='Homeruns', index='yearID', columns='teamID', aggfunc='sum', fill_value=0.0)
team_homeruns_per_year_df.head()
```

```python
print('Plot as a heatmap')
plt.rcParams["figure.figsize"] = (10, 10)
team_homeruns_per_year_df_t = team_homeruns_per_year_df.T
_ = sns.heatmap(team_homeruns_per_year_df_t)
```

```python
print('Plot as a heatmap with annotate each cell')
_ = sns.heatmap(team_homeruns_per_year_df_t, annot=True, fmt='d')
```

```python
print('Plot as a heatmap with OAK 2008 as "center"')
_ = sns.heatmap(team_homeruns_per_year_df_t, center=team_homeruns_per_year_df_t.loc['OAK', 2008])
```

```python
print('Plot as a heatmap with NYA 2008 as "center"')
_ = sns.heatmap(team_homeruns_per_year_df_t, center=team_homeruns_per_year_df_t.loc['NYA', 2008])
```

```python
print('Plot as a clustermap (reformat the heatmap so similar rows are next to each other)')
_ = sns.clustermap(team_homeruns_per_year_df_t)
```

```python
print('Plot as a clustermap custerring only teams')
_ = sns.clustermap(team_homeruns_per_year_df_t, col_cluster=False)
```
